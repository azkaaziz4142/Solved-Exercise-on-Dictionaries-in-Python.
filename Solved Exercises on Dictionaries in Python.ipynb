```python
# PART A

info={'name':'jhon','age':'32','city':'boston','hobbies':'reading'}

# PART B
def word_frequency(text):
    text = text.lower()
    text = ''.join(e for e in text if e.isalnum() or e.isspace())
    words = text.split()
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    
    return word_count

# PART C
inventory = {
    "apples": 50,
    "bananas": 30,
    "oranges": 20,
    "grapes": 15
}
inventory["mangoes"] = 25
inventory["bananas"] -= 5
print(inventory)
item = "oranges"
if item in inventory:
    print(f"Inventory for {item}: {inventory[item]}")
else:
    print(f"{item} not found in the inventory.")

# PART D
orders = [{"customer_id": 1, "product": "laptop", "category": "electronics"},{"customer_id": 2, "product": "phone", "category": "electronics"},{"customer_id": 3, "product": "shirt", "category": "clothing"}, {"customer_id": 4, "product": "headphones", "category": "electronics"}]
category_count = {}
for order in orders:
    category = order["category"]
    if category in category_count:
        category_count[category] += 1
    else:
        category_count[category] = 1
most_popular_category = max(category_count, key=category_count.get)
print(f"The most popular category is: {most_popular_category}") 

# PART E
student_grades = {"Alice": 85, "Bob": 92,"Charlie": 78,"David": 88,"Eve": 95}
def calculate_average(grades):
    total = sum(grades.values())
    num_students = len(grades)
    average = total / num_students
    return average
average_grade = calculate_average(student_grades)
print("The average grade is:", average_grade)

# PART F 
inventory = {}
def add_product(name, quantity, price):
    if name in inventory:
        print(f"Product {name} already exists. Use update_product to change quantity.")
    else:
        inventory[name] = {'quantity': quantity, 'price': price}
        print(f"Product {name} added successfully.")
def update_product(name, quantity):
    if name in inventory:
        inventory[name]['quantity'] += quantity
        print(f"Product {name} quantity updated by {quantity}.")
    else:
        print(f"Product {name} does not exist. Please add it first.")
def calculate_inventory_value():
    total_value = 0
    for product in inventory.values():
        total_value += product['quantity'] * product['price']
    return total_value
def generate_low_stock_report(threshold):
    low_stock_items = []
    for product_name, product_info in inventory.items():
        if product_info['quantity'] <= threshold:
            low_stock_items.append((product_name, product_info['quantity']))
    
    if low_stock_items:
        print("Low-Stock Items:")
        for item in low_stock_items:
            print(f"{item[0]}: {item[1]} units left")
    else:
        print("No low-stock items.")
    
# Example usage

# Add some products to the inventory
add_product("Laptop", 10, 800)
add_product("Smartphone", 5, 600)
add_product("Headphones", 50, 50)

# Update product quantities
update_product("Laptop", 5)  # Add 5 more laptops
update_product("Smartphone", -3)  # Sell 3 smartphones

# Calculate the total inventory value
total_value = calculate_inventory_value()
print(f"The total inventory value is: ${total_value}")

# Generate a low-stock report (threshold is 10 units)
generate_low_stock_report(10)

# PART G

students = {
    "Alice": {"Math": 85, "English": 92, "History": 78},
    "Bob": {"Math": 88, "English": 79, "History": 85},
    "Charlie": {"Math": 90, "English": 84, "History": 91},
    "David": {"Math": 78, "English": 88, "History": 80},
    "Eve": {"Math": 95, "English": 93, "History": 89}
}

# Function to calculate the average grade for each student
def calculate_average(student_grades):
    total = sum(student_grades.values())
    num_subjects = len(student_grades)
    return total / num_subjects

# Function to find the highest and lowest grades in a specific subject
def find_highest_lowest_grade(subject):
    highest_grade = None
    lowest_grade = None
    highest_student = ""
    lowest_student = ""
    
    for student, grades in students.items():
        if subject in grades:
            grade = grades[subject]
            if highest_grade is None or grade > highest_grade:
                highest_grade = grade
                highest_student = student
            if lowest_grade is None or grade < lowest_grade:
                lowest_grade = grade
                lowest_student = student
    
    return highest_student, highest_grade, lowest_student, lowest_grade

# Function to sort students by their average grades
def sort_students_by_average():
    student_averages = {}
    for student, grades in students.items():
        student_averages[student] = calculate_average(grades)
    
    sorted_students = sorted(student_averages.items(), key=lambda x: x[1], reverse=True)
    
    return sorted_students

# Example usage

# Calculate and print average grade for each student
print("Average grades for each student:")
for student, grades in students.items():
    avg = calculate_average(grades)
    print(f"{student}: {avg:.2f}")

# Find and print the highest and lowest grades in Math
subject = "Math"
highest_student, highest_grade, lowest_student, lowest_grade = find_highest_lowest_grade(subject)
print(f"\nHighest grade in {subject}: {highest_student} with {highest_grade}")
print(f"Lowest grade in {subject}: {lowest_student} with {lowest_grade}")

# Sort students by average grades and print the sorted list
print("\nStudents sorted by average grade:")
sorted_students = sort_students_by_average()
for student, avg in sorted_students:
    print(f"{student}: {avg:.2f}")

# PART H
products = {
    "Laptop": {"price": 800, "quantity": 50},
    "Smartphone": {"price": 600, "quantity": 100},
    "Headphones": {"price": 50, "quantity": 200},
    "Tablet": {"price": 300, "quantity": 75},
}

cart = {}

# Function to add a product to the cart
def add_to_cart(product_name, quantity):
    if product_name in products:
        # Check if enough stock is available
        if products[product_name]["quantity"] >= quantity:
            # If the product is already in the cart, update the quantity
            if product_name in cart:
                cart[product_name]["quantity"] += quantity
            else:
                cart[product_name] = {"price": products[product_name]["price"], "quantity": quantity}
            # Reduce stock from the products dictionary
            products[product_name]["quantity"] -= quantity
            print(f"{quantity} {product_name}(s) added to the cart.")
        else:
            print(f"Sorry, only {products[product_name]['quantity']} {product_name}(s) are available in stock.")
    else:
        print(f"Product {product_name} not found.")

# Function to remove a product from the cart
def remove_from_cart(product_name, quantity):
    if product_name in cart:
        if cart[product_name]["quantity"] >= quantity:
            cart[product_name]["quantity"] -= quantity
            # Add the removed quantity back to stock
            products[product_name]["quantity"] += quantity
            if cart[product_name]["quantity"] == 0:
                del cart[product_name]  # Remove product completely if quantity is 0
            print(f"{quantity} {product_name}(s) removed from the cart.")
        else:
            print(f"Error: You have only {cart[product_name]['quantity']} {product_name}(s) in the cart.")
    else:
        print(f"Product {product_name} is not in your cart.")

# Function to calculate the total cost of items in the cart
def calculate_total():
    total_cost = 0
    for item, details in cart.items():
        total_cost += details["price"] * details["quantity"]
    return total_cost

# Function to apply discount to the total cost
def apply_discount(total_cost, discount_percent):
    discount_amount = (discount_percent / 100) * total_cost
    return total_cost - discount_amount

# Function to apply tax to the total cost
def apply_tax(total_cost, tax_percent):
    tax_amount = (tax_percent / 100) * total_cost
    return total_cost + tax_amount

# Example usage
add_to_cart("Laptop", 2)  # Adding 2 laptops
add_to_cart("Smartphone", 1)  # Adding 1 smartphone
add_to_cart("Headphones", 3)  # Adding 3 headphones

# Display the cart
print("\nYour Cart:")
for product, details in cart.items():
    print(f"{product}: {details['quantity']} x ${details['price']}")

# Calculate and display the total cost
total_cost = calculate_total()
print(f"\nTotal cost before discount and tax: ${total_cost:.2f}")

# Apply discount and tax
discount_percent = 10  # 10% discount
tax_percent = 8  # 8% tax

total_after_discount = apply_discount(total_cost, discount_percent)
total_after_tax = apply_tax(total_after_discount, tax_percent)

print(f"\nTotal cost after {discount_percent}% discount: ${total_after_discount:.2f}")
print(f"Total cost after {tax_percent}% tax: ${total_after_tax:.2f}")

# Remove a product from the cart
remove_from_cart("Headphones", 2)  # Removing 2 headphones

# Final cart and total
print("\nFinal Cart:")
for product, details in cart.items():
    print(f"{product}: {details['quantity']} x ${details['price']}")

final_total = calculate_total()
final_total_after_discount = apply_discount(final_total, discount_percent)
final_total_after_tax = apply_tax(final_total_after_discount, tax_percent)

print(f"\nFinal total cost after discount and tax: ${final_total_after_tax:.2f}")




```
